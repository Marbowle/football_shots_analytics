```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data_path = "D:/football_shots_analytics/outputs/tracks_players_sort.csv"
```

```python
df = pd.read_csv(data_path)
```

```python
df.head(10)
```

```python
unique_ids = df['track_id'].unique()
subset_ids = [0,1,2,3,4,5,6,7]
subset_df = df[df['track_id'].isin(subset_ids)]
```

```python
fig = plt.figure(figsize=(12,8))
for track_id in subset_ids:
    track = subset_df[subset_df['track_id'] == track_id]
    plt.plot(track['cx'], track['cy'], label=f"Player {track_id}")
    plt.scatter(track['cx'].iloc[0], track['cy'].iloc[0], color='green', label='Start', marker='o')
    plt.scatter(track['cx'].iloc[-1], track['cy'].iloc[-1], color='red', label='Stop', marker='x')

plt.gca().invert_yaxis()

plt.title("Trajektoria zawodnik√≥w")
plt.xlabel("Pozycja x w pikselach")
plt.ylabel("Pozycja y w pikselach")
plt.legend(loc='upper left')
plt.show()
```
