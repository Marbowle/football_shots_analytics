# Create players tracking with ByteTrack


```python
df = pd.read_csv("D:/football_shots_analytics/outputs/detections_players_for_tracking.csv")
```

```python
df.head(10)
```

```python
df = df.sort_values(by=['frame']).reset_index(drop=True)
```

```python
df['cx'] = round((df['x1'] + df['x2']) / 2, 4)
df['cy'] = round((df['y1'] + df['y2']) / 2, 4)
```

```python
df.head(10)
```

```python
active_tracker = {}
next_track_id = 0
distance_threshold = 70
track_result = []
```

```python
def find_closest_track(cx, cy, active_tracker):
    nearest_id = None
    nearest_dist = float('inf')
    for track_id, info in active_tracker.items():
        dist = math.sqrt((cx - info['cx']) ** 2 + (cy - info['cy']) ** 2)
        if dist < nearest_dist:
            nearest_dist = dist
            nearest_id = track_id
    return nearest_id, nearest_dist
```

```python
for frame in sorted(df['frame'].unique()):
    current_detections = df[df['frame'] == frame]
    for _, detection in current_detections.iterrows():
        cx, cy = detection['cx'], detection['cy']
        nearest_id, nearest_distance = find_closest_track(cx, cy, active_tracker)

        if nearest_distance < distance_threshold:
            track_id = nearest_id
            active_tracker[track_id]['cx'] = cx
            active_tracker[track_id]['cy'] = cy
            active_tracker[track_id]['last_frame'] = frame
        else:
            track_id = next_track_id
            active_tracker[track_id] = {'cx': cx, 'cy': cy, 'last_frame': frame}
            next_track_id += 1

        track_result.append({
            'frame': frame,
            'track_id': track_id,
            'cx': cx,
            'cy': cy,
            'x1': detection['x1'],
            'y1': detection['y1'],
            'x2': detection['x2'],
            'y2': detection['y2'],
            'confidence': detection['confidence'],
        } )

    for track_id in list(active_tracker.keys()):
        if frame - active_tracker[track_id]['last_frame'] > 10:
            del active_tracker[track_id]
```

```python
track_df = pd.DataFrame(track_result)
```

```python
track_df.head(10)
```

```python
track_df.to_csv("D:/football_shots_analytics/outputs/tracks_players.csv", index=False)
```
