```python
from enum import unique

import pandas as pd
import math
```

```python
data_path = "D:/football_shots_analytics/outputs/tracks_players_sort.csv"
```

```python
df = pd.read_csv(data_path)
```

```python
df.head()
```

```python
distance_summary = []
df['cx']= df['cx'].astype(float)
df['cy'] = df['cy'].astype(float)
```

```python
df = df.sort_values(['track_id', 'frame', 'confidence'], ascending=[True, True, False])
df = df.drop_duplicates(subset=['track_id', 'frame'], keep='first')
df = df.reset_index(drop=True)
```

```python
for track_id in df['track_id'].unique():
    subset = df[df['track_id'] == track_id].sort_values('frame')

    total_distance = 0
    for i in range(1, len(subset)):
        x1, y1 = subset.iloc[i -1][['cx', 'cy']]
        x2, y2 = subset.iloc[i][['cx', 'cy']]
        dx = x2 - x1
        dy = y2 - y1
        dist = math.sqrt(dx**2 + dy**2)
        total_distance += dist

    distance_summary.append({
        'track_id': track_id,
        'total_distance_px': total_distance,
        'frames_tracked': len(subset)
    })
```

```python
print(distance_summary[:5])
```

```python
df_distance_summary = pd.DataFrame(distance_summary)
```
